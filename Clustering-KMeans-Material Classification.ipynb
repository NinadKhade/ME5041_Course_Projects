```python
import pandas as pd
```

```python
dataset = {
    'Material': ['Zinc', 'Magnesium', 'Molybdenum', 'Rhodium', 'Graphite', 'Germanium', 'Boron', 'Sapphire'],
    'Thermal Conductivity': [116, 156, 138, 150, 25, 60, 27, 35],
    'Youngs Modulus': [108, 45, 330, 275, 10, 103, 320, 345],
}
```

```python
df = pd.DataFrame(dataset)
df
```

## Clustering of dataset

### Prepare data

```python
X = df[['Thermal Conductivity', 'Youngs Modulus']]
X
```

```python
from sklearn.preprocessing import MinMaxScaler

X_scaled = MinMaxScaler().fit_transform(X)
X_scaled
```

### Visualize data

```python
import matplotlib.pyplot as plt

plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.xlabel('Thermal Conductivity')
plt.ylabel('Youngs Modulus')
plt.show()
```

### Implement clustering

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0).fit(X_scaled)
```

```python
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

### Visualize cluster

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200, c='red')
plt.xlabel('Thermal Conductivity')
plt.ylabel('Youngs Modulus')
plt.show()
```
